```python
file_name = "telco_updated_data_vr2_jannat.xlsx"  # Replace with the actual filename
# Read the Excel file into a DataFrame
df = pd.read_excel(file_name)
# Now you can work with the DataFrame 'df'
print(df.head())  # Display the first few rows of the DataFrame
df['Format Start Date'] = pd.to_datetime(df['Format Start Date'])
df['Format End Date'] = pd.to_datetime(df['Format End Date'])
df['Primary Text (caption)']=df['Primary Text (caption)'].apply(str)
def delete_text_between_patterns(text):
    # Replace text between <br> and </br> with an empty string
    text = re.sub(r'<br />.*?<br />', '', text)
    # Replace text between <a> and </a> with an empty string
    text = re.sub(r' <a.*?</a>', '', text)
    return text
# Apply the function to the specified column
df['Primary Text (caption)'] = df['Primary Text (caption)'].apply(delete_text_between_patterns)
df['Primary Text (caption)'] = df['Primary Text (caption)'].replace({'<br />': ''}, regex=True)
df['Primary Text (caption)'] = df['Primary Text (caption)'].replace({'&quot;': ''}, regex=True)
distinct_df =df.drop_duplicates(subset=['Primary Text (caption)'], keep='last')
print(distinct_df.shape)
# Assume df is your DataFrame
# Save DataFrame to an Excel file
distinct_df.to_excel('final.xlsx', index=False)
```
