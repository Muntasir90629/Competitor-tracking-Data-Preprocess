```python
import pandas as pd
import glob
import os
```

```python
# Define the folder path
folder_path = r"C:\Users\MdMuntasirulHoque\OneDrive - ADA Global\Desktop\Ad library"

# Get all CSV files from the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
```

```python
# Read and combine all CSV files
df_list = [pd.read_csv(file, low_memory=False) for file in csv_files]
final_df = pd.concat(df_list, ignore_index=True)
```

```python
final_df.shape
```

```python
# Define the required columns and new names
column_mapping = {
    "ad_archive_id": "Ad Archive ID",
    "page_name": "Page name",
    "publisher_platform/0": "Platform_0",
    "publisher_platform/1": "Platform_1",
    "publisher_platform/2": "Platform_2",
    "publisher_platform/3": "Platform_3",
    "snapshot/body/text": "Primary Text (caption)",
    "snapshot/caption": "Preview URL (Button URL)",
    "snapshot/cards/0/original_image_url": "Original Image",
    "snapshot/cards/0/resized_image_url": "Resized Image",
    "snapshot/cards/0/title": "Headline",
    "snapshot/cards/0/video_hd_url": "Video 1 HD",
    "snapshot/cards/0/video_sd_url": "Video 1 sd",
    "snapshot/cards/1/body": "Primary Text (caption)",
    "snapshot/cta_text": "CTA TEXT",
    "snapshot/display_format": "Creative Format",
    "snapshot/link_url": "Destination Url",
    "snapshot/title": "Headline2",
    "start_date": "Start Date"
}

# Select only the required columns
selected_df = final_df[list(column_mapping.keys())]

# Rename the columns
selected_df = selected_df.rename(columns=column_mapping)

# Drop duplicate rows
selected_df = selected_df.drop_duplicates()
```
